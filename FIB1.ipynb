```python
amount = 100000
rate = 0.05
total_interest_added = 0

def calculate_interest(num_of_days, annual_interest_rate, total_amount):
    return total_amount * (1 + annual_interest_rate / 365) ** num_of_days - total_amount

# October calculations
total_interest_added += calculate_interest(5, rate, amount)
amount += calculate_interest(5, rate, amount)
amount += 17000
total_interest_added += calculate_interest(12-5, rate, amount)
amount += calculate_interest(12-5, rate, amount)
amount += 35000
total_interest_added += calculate_interest(21-12, rate, amount)
amount += calculate_interest(21-12, rate, amount)
amount -= 55000
total_interest_added += calculate_interest(31-21, rate, amount)
amount += calculate_interest(31-21, rate, amount)
print("Interest added for October: ", total_interest_added)
print("Balance at the end of the month: ", amount)

# Reset total interest added for November calculations
total_interest_added = 0
total_interest_added += calculate_interest(9, rate, amount)
amount += calculate_interest(9, rate, amount)
amount -= 25000
total_interest_added += calculate_interest(30-9, rate, amount)
amount += calculate_interest(30-9, rate, amount)
amount += 100000
print("Interest added for November: ", total_interest_added)
print("Balance at the end of the month: ", amount)
```

```python
import math

def nominal_interest_rate(inflation_rate = 0.1):
    r_yearly = inflation_rate
    r_half_yearly = 2 * ((1 + inflation_rate)**(1/2) - 1)
    r_quarterly = 4 * ((1 + inflation_rate)**(1/4) - 1)
    r_monthly = 12 * ((1 + inflation_rate)**(1/12) - 1)
    r_perma = math.log(1 + inflation_rate)

    return {
            "Yearly": r_yearly,
            "Half-yearly": r_half_yearly,
            "Quarterly": r_quarterly,
            "Monthly": r_monthly,
            "Continuous": r_perma
        }

nominal_interest_rate()
```

```python
import numpy as np

def monthly_interest_payout(principal, annual_rate, months):
    monthly_rate = annual_rate / 12
    total_interest = (monthly_rate * principal) * months
    return total_interest

def compound_interest(principal, annual_rate, months):
    monthly_rate = annual_rate / 12
    total_amount = principal * (1 + monthly_rate) ** months
    total_interest = total_amount - principal
    return total_interest

def effective_interest_rate(monthly_rates):
    product = np.prod([1 + rate / 12 for rate in monthly_rates])
    eir = product - 1
    return eir

principal = 10000
annual_rate = 0.06 
months = 12

bank_A_rates = [i / 100 for i in range(1, 13)]
bank_B_rate = 6.5 / 100
bank_B_rates = [bank_B_rate] * 12 
bank_C_rates = [i / 100 for i in range(12, 0, -1)]

eir_A = effective_interest_rate(bank_A_rates)
eir_B = effective_interest_rate(bank_B_rates)
eir_C = effective_interest_rate(bank_C_rates)

# efektywne stopy procentowe
print(f"Efektywna stopa procentowa Banku A: {eir_A:.4%}")
print(f"Efektywna stopa procentowa Banku B: {eir_B:.4%}")
print(f"Efektywna stopa procentowa Banku C: {eir_C:.4%}")

interest_no_compound = monthly_interest_payout(principal, annual_rate, months)
interest_with_compound = compound_interest(principal, annual_rate, months)

print(f"Odsetki przy miesięcznych wypłatach (bez kapitalizacji): {interest_no_compound:.2f} PLN")
print(f"Odsetki przy miesięcznej kapitalizacji: {interest_with_compound:.2f} PLN")
```
